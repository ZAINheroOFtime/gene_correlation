```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
```

```python
gc_table = pd.read_csv('latent_immunomodulators_gene_combo_filtered.csv')
```

```python
gc_table
```

```python
emptygenes = []
```

```python
def calc_corr(row):
    directory = "VAE files\\" + row['tissue'] +'\\'
    gene_combo = row['gene_combo'].split('_')
    gene1 = np.load(directory+gene_combo[0]+'.npy')
    gene2 = np.load(directory+gene_combo[1]+'.npy')
    if gene1.flatten().sum()==0:
        if [gene_combo[0], row['tissue']] not in emptygenes:
            emptygenes.append([gene_combo[0], row['tissue']])
    if gene2.flatten().sum()==0:
          if [gene_combo[1], row['tissue']] not in emptygenes:
            emptygenes.append([gene_combo[1], row['tissue']])      
    corr, pval = pearsonr(gene1.flatten(), gene2.flatten())
    return pd.Series([corr,pval])
```

```python
gc_table[["correlation", "corr_pval"]]  = gc_table.apply(calc_corr, axis=1)
```

```python
emptygenes_df = pd.DataFrame(emptygenes, columns = ['Gene', 'Tissue'])
```
